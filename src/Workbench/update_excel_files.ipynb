```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os

sns.set(style="darkgrid")
pd.set_option('display.max_rows', None)
```

```python
# INPUT_DIR = '/Users/jamesbamber/Documents/Projects/FSPL (UBC)/Bamber_Logisymetrix/Individual process data sets/'
# OUTPUT_DIR = '/Users/jamesbamber/Documents/Projects/FSPL (UBC)/Bamber_Logisymetrix/Individual process data sets (updated)/'
INPUT_DIR = "C:\\Users\\Grass_Valley\\Documents\\Projects\\FSPL (UBC)\\GitLab\\openlca\\resources\\Individual process data sets\\"
OUTPUT_DIR = "C:\\Users\\Grass_Valley\\Documents\\Projects\\FSPL (UBC)\\GitLab\\openlca\\resources\\Individual process data sets (updated)\\"

with os.scandir(INPUT_DIR) as entries:
    for entry in entries:
        if entry.is_file() and entry.name[:6] == "Farm__":
            farm_filename = entry.name
            print(farm_filename)
            in_farm_filepath = INPUT_DIR + farm_filename
            manure_df = pd.read_excel(in_farm_filepath, sheet_name="manure")
            if "Allocation factor" not in manure_df.columns:
                manure_df["Allocation factor"] = np.NaN

            production_df = pd.read_excel(in_farm_filepath, sheet_name="production")
            if "Allocation factor" not in production_df.columns:
                production_df["Allocation factor"] = np.NaN
                production_df.at[13, "Allocation factor"] = 0.909102060359958
                production_df.at[15, "Allocation factor"] = 0.0908979396400422

            out_farm_filepath = OUTPUT_DIR + farm_filename
            with pd.ExcelWriter(out_farm_filepath) as writer:
                manure_df.to_excel(writer, sheet_name="manure")  
                production_df.to_excel(writer, sheet_name="production")
            break
```
